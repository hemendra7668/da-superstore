```python
import pandas as pd
import numpy as np
salesdata= {"person":["rajesh","ravi","pradeep","amit","rakesh","suresh","raj","raju"],
            "company":["com a","com a","com a","com b", "com b","com c","com a","com a"],
            "Sales" : [ 250 , 250 ,150 , 340 , np.nan, 243 ,350 , 250],
            "Age" : [ 23 , 24, 25, 250 , np.nan, 27, 24 , 24] }
```

handling the missing the missing value:--

```python
salesdf=pd.DataFrame(salesdata)
salesdf
```

clean data

```python
# removing the missing values
salesdf['Age'].replace(250,25,inplace=True)
avg_age=int(salesdf['Age'].mean())
avg_age
salesdf['Age'].fillna(avg_age,inplace=True)
salesdf
```

```python
salesdf['Age'].replace(56,28, inplace=True)
salesdf
```

```python
salesdf['Sales'].fillna(0, inplace=True)
salesdf
```

```python
salesdf['Age'].drop_duplicates(inplace=True)
salesdf
```
